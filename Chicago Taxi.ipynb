```python
import pandas as pd

df = pd.read_csv('chicago_taxi.csv')
df
```

```python
df_lat_log = df.iloc[ : , 17:19]
df_lat_log
```

```python
df_lat_log = df_lat_log.fillna(df_lat_log.mean())
df_lat_log
```

```python
import matplotlib.pyplot as plt
plt.scatter(df_lat_log['Pickup Centroid Latitude'], df_lat_log['Pickup Centroid Longitude'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

```python
X=df_lat_log[['Pickup Centroid Latitude', 'Pickup Centroid Longitude']].values
X
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaler.inverse_transform(kmeans.cluster_centers_)
```

```python
labels=kmeans.predict(X)

plt.scatter(X[:,1], X[:,0], c= labels, s=25, alpha=0.25)
plt.xlabel('Longitude')
plt.ylabel('Latitude')

centroid = kmeans.cluster_centers_
# centroid = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centroid[:,1], centroid[:,0], s=50, color='black')

plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(df_lat_log)
```

```python
centers=kmeans.cluster_centers_
```

```python
kmeans_inertia = pd.DataFrame(data=[], index=(2,11), columns=['inertia'])

for n_clusters in range(2,11):
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(df_lat_log)
    kmeans_inertia.loc[n_clusters]=kmeans.inertia_
    
kmeans_inertia
```

```python
kmeans_inertia.plot(kind='line', y='inertia')
plt.xlabel('No of clusters')
plt.show()
```
